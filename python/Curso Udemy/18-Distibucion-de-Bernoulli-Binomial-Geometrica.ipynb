```python
Las funciones de Bernoulli estan en el paquete scipy.stat.Bernoulli
pmf(k,p) -probabilidad de funcion
cdf(k,p) - Acumulada
ppd(q,p)  -funcion de probabilidad
rvs(p,size) - generador de numeros aleatorios

donde p es la probabilidad e exito
```

```python
from scipy.stats import bernoulli
import matplotlib.pyplot as plt
```

```python
p=0.7
```

```python
mean, var, skew,kurt=bernoulli.stats(p, moments = 'mvsk')
print("Media %f" %mean)
print("Varianza %f" %var)
print("Sesgo %f" %skew)
print("Curtosis %f" %kurt)
```

```python
#mostar la funcion de probabilidad
fix,ax = plt.subplots(1,1)
x = bernoulli.rvs(p,size=1000)
ax.hist(x)
plt.show()
```

##  Binomial


Las funciones Binomiales estan en el paquete scipy.stat.Binom

pmf(k,n,p) -probabilidad de funcion

cdf(k,n,p) -Acumulada

ppd(q,n,p)  -funcion de probabilidad

rvs(p,n,size) - generador de numeros aleatorios

donde p es la probabilidad e exito

```python
from scipy.stats import binom
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig,ax = plt.subplots(1,1)
n=7
p=0.4

mean, var , skew, kurt = binom.stats(n,p,moments='mvsk')

print("Media %f" %mean)
print("Varianza %f" %var)
print("Sesgo %f" %skew)
print("Curtosis %f" %kurt)


x=np.arange(0,n+1)
ax.plot(x,binom.pmf(x,n,p),'bo',ms=8,label = "Funcion de densidad de B(7,0,4")
ax.vlines(x,0,binom.pmf(x,n,p),colors='b',lw=4, alpha=0.5)

rv=binom(n,p)
ax.vlines(x,0,rv.pmf(x),colors='k', linestyle='--',lw=1, label="Distribucion teorica")
ax.legend(loc='best',frameon=False)
plt.show()
```

```python
fig,ax = plt.subplots(1,1)
n=7
p=0.4

mean, var , skew, kurt = binom.stats(n,p,moments='mvsk')

print("Media %f" %mean)
print("Varianza %f" %var)
print("Sesgo %f" %skew)
print("Curtosis %f" %kurt)


x=np.arange(0,n+1)
ax.plot(x,binom.cdf(x,n,p),'bo',ms=8,label = "Funcion de densidad de B(7,0,4")
ax.vlines(x,0,binom.cdf(x,n,p),colors='b',lw=4, alpha=0.5)

rv=binom(n,p)
ax.vlines(x,0,rv.pmf(x),colors='k', linestyle='--',lw=1, label="Distribucion teorica")
ax.legend(loc='best',frameon=False)
plt.show()

fix,ax=plt.subplots(1,1)
r= binom.rvs(n,p,size=10000)
ax.hist(r, bins=n)
plt.show()
```

## Distribucion geometrica

X es el nuemro de intentos hasta conseguir el primer exito

Las funciones Binomiales estan en el paquete scipy.stat.geom

pmf(k,n,p) -probabilidad de funcion

cdf(k,n,p) -Acumulada

ppd(q,n,p)  -funcion de probabilidad

rvs(p,n,size) - generador de numeros aleatorios

donde p es la probabilidad e exito

```python
from scipy.stats import geom
```

```python
fig, ax=plt.subplots(1,1)
p=0.3

mean , var , skew, kurt =geom.stats(p,moments='mvsk')
print("Media %f" %mean)
print("Varianza %f" %var)
print("Sesgo %f" %skew)
print("Curtosis %f" %kurt)

x = np.arange(geom.ppf(0.001,p),geom.ppf(0.99,p))
ax.plot(x,geom.pmf(x,p),'bo',ms=8,label = "Funcion de probalidad de Geom 0.3")
ax.vlines(x,0,geom.pmf(x,p),colors='b', lw=4, alpha = 0.5)
rv = geom(p)
ax.vlines(x,0,rv.pmf(x),colors='k', linestyles="--",lw=1,label=" Frozen PMF")
ax.legend(loc='best')
plt.show()

fig, ax=plt.subplots(1,1)
prob=geom.cdf(x,p)
ax.plot(x,prob,'bo',ms=8,label = "Funcion de distribucion acumulada")
plt.show()

fix,ax=plt.subplots(1,1)
r=geom.rvs(p,size=10000)
plt.hist(r)
```
