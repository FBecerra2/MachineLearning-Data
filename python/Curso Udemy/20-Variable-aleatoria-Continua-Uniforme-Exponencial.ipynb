## Distribucion Uniforme

Elegir un elemento del intervalo de manera equiprobable (AI)

scipy.stats.Uniform

pmf(k,loc,scale)

cdf(k,loc,scale)

ppf(q,loc,scale)

rvs(n,loc,scale)

Donde la distribucion Uniforme esta definida ne el intervalo [loc, loc+scale]



```python
from scipy.stats import uniform
import matplotlib.pyplot as plt
import numpy as np

a=0
b=1
loc = a
scale = b-a

fig,ax = plt.subplots(1,1)
rv = uniform(loc=loc,scale=scale)
mean, var , skew, kurt = rv.stats(moments='mvsk')
print("Media %f" %mean)
print("Varianza %f" %var)
print("Sesgo %f" %skew)
print("Curtosis %f" %kurt)
x = np.linspace(-0.1,1.1,120)
ax.plot(x,rv.pdf(x),'k-', lw=2,label="Uniforme en el intervalo U(0,1)")

r=rv.rvs(size=100000)
ax.hist(r,density=True,histtype="stepfilled", alpha=0.25)
ax.legend(loc='best', frameon=False)
plt.show()
```

## Distribucion Exponencial

El tiempo que falta para que ocurra algo, Tiempo en que pase un colectivo

scipy.stats.expon

pmf(k,scale)
cdf(k,scale)
ppf(q,scale)
rvs(n,scale)

Donde scale es 1/lambda es la inversa del tiempo que ocurre entre dos suceso de la distribucion.

```python
from scipy.stats import expon
```

```python
fig,ax = plt.subplots(1,1)

lam=10

rv= expon(scale=1/lam)

mean,var,skew,kurt = rv.stats(moments='mvsk')
print("Media %f" %mean)
print("Varianza %f" %var)
print("Sesgo %f" %skew)
print("Curtosis %f" %kurt)

x=np.linspace(0,2,1000)
ax.plot(x,rv.pdf(x),'r-',lw=5,alpha=0.6, label="Exp(10")
r = rv.rvs(size=100000)
ax.hist(r,density = True, histtype='stepfilled',alpha=0.2)
ax.legend(loc="best", frameon=False)
```
