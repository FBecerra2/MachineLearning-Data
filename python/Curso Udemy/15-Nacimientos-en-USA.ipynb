```python
quartiles=np.percentile(births['births'],[25,50,75])
```

```python
med = quartiles[1] #mediana
IQR = quartiles[2] - quartiles[0]
sig = 0.75*IQR #desviacion tipica
```

```python
births = births.query("(births > @med - 5*@sig) & (births < @med + 5*@sig)")
births.shape

#Esta es una forma de sacar la mediana y desviacion tipica se usa en Machine Learning cuando no es posible calcularlo de forma explicita (Elimina factores extraños)
```

```python
births['day'] = births['day'].astype(int) #Cambiar numeros a enteres
```

```python
births.head()
```

```python
births.index = pd.to_datetime(10000*births.year+100*births.month+births.day, format="%Y%m%d") # Escribir la fecha de una manera que lo interprete Python
```

```python
births.head()
```

```python
births['dayofweek'] = births.index.dayofweek #Agregar el dia de la semana
```

```python
births.head()
```

```python
births.pivot_table('births',index='dayofweek',columns='decade',aggfunc='mean').plot()
plt.gca().set_xticklabels(["","Lun","Mar","Mie","Jue","Vie", "Sab","Dom"])
plt.ylabel("Medida de nacimientos por dia")
plt.show
```

```python
births_by_date = births.pivot_table('births',[births.index.month, births.index.day],aggfunc="mean") #Dias raros para tener hijos
```

```python
births_by_date.head()
```

```python
births_by_date.index = [pd.datetime(2020,month,day)
for (month,day) in births_by_date.index]
```
