# Manipulacion de data frames

```python
df = pd.DataFrame({
    'key' : ['A','B','C','A','B','C'],
    'values1': range(6),
    'values2': np.random.randint(0,100,6)
})
```

```python
df
```

# Agregate

```python
df.groupby('key').aggregate(['min', np.median, np.mean,max])
```

```python
df.groupby('key').aggregate({
    'values1':'min','values2':'max'})
```

## Filter

```python
df.groupby('key').std() #Desviacion standard
```

```python
def filter_func(x):
    return x['values2'].std()>3
```

```python
display('df',"df.groupby('key').std()","df.groupby('key').filter(filter_func)")
```

```python
df.groupby('key').filter(filter_func)
```

# Transform

```python
df.groupby('key').transform(lambda x: x - x.mean())
```

# Apply

```python
def norm_by_col2(x):
    x['values1'] /= x['values2'].sum()
    return x
```

```python
df.groupby('key').apply(norm_by_col2)
```

# Division por columnas

```python
L = [0,1,0,1,2,0] #Cada uno de los numeros de la list alos agrupa la posicion 1, 3 y 6 grupo 0 - posicion 2 y 4 grupo 1 - posicion 5 grupo 2
print(df)
df.groupby(L).sum() 
#Suma los que estan en la posicion que indica el grupo 0 = posicion 0,2,5 con valores 0+2+5=7(values1) 
```

```python
df2=df.set_index('key')
df2
```

```python
mapping = {
    'A':'vowel',
    'B':'consonant',
    'C':'consonant'
} # Cambia los grupos a los nuevos nombres
```

```python
df2.groupby(mapping).sum()
```

```python
df2.groupby(str.lower).mean() #pasa el nombre de las claves a minisculas
```

```python
df2.groupby([str.lower, mapping]).mean()
```

# Graficos desde data frames y exportacion

```python
years = [year for year in range(1900,2020)]
deads = [(y+ np.random.uniform(0,100)-1850) for y in years]
```

```python
df = pd.DataFrame({
    "year": years,
    "deads": deads
})
```

```python
df.head()
```

```python
df.plot(kind="scatter",x="year", y = "deads", figsize=(10,8))
```

# Exportar Grafico

```python
my_plot=df.plot(x="year", y = "deads", figsize=(10,8))
```

```python
my_fig=my_plot.get_figure()
```

```python
my_fig.savefig("dataframe/data_frame_plot.png")
```
