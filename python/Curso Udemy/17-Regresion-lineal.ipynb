# Diabetes

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error, r2_score
```

```python
diabetes= datasets.load_diabetes()
type(diabetes)
```

```python
type(diabetes.data)
```

```python
diabetes.data.shape
```

```python
x = diabetes.data[:,np.newaxis,2]
x
```

```python
y = diabetes.target
```

```python
x_train = x[:-60]
x_test=x[-60:]
y_train= y[:-60]
y_test= y[-60:]
```

```python
mod = linear_model.LinearRegression()
```

```python
mod.fit(x_train, y_train)
```

```python
mod.coef_ #coeficiente de las X
```

```python
mod.intercept_#Orednada en el origen
```

$$\tilde y = 949.43526038\cdot x + 152.13348416289594$$

```python
plt.scatter(x_train,y_train,color='black')
plt.plot(x_train, mod.predict(x_train),color='blue', linewidth=3)#Darle las x iguales a scatter -- mod.predic es para predecir los puntos
plt.show()
```

$$MSE = \frac{\sum_{i=1}^n (y_i - \tilde{y_i})^2}{n}$$

```python
mean_squared_error(y_train,mod.predict(x_train))#R medio o valor medio cuadrado
```

```python
r2_score(y_train,mod.predict(x_train))
```

# Entrenemiento y validacion de un modelo lineal (Testing)

```python
y_pred = mod.predict(x_test)
```

```python
mean_squared_error(y_test, y_pred)
```

```python
r2_score(y_test, y_pred)
```

```python
plt.scatter(x_train,y_train,color='black')
plt.scatter(x_test,y_test,color='red')
plt.plot(x_train, mod.predict(x_train),color='blue', linewidth=3)#Darle las x iguales a scatter -- mod.predic es para predecir los puntos
plt.show()
```
