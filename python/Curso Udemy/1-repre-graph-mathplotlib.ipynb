```python
def f(x):
    return np.exp(-x)*np.cos(2*np.pi*x)
```

```python
x1 = np.arange(0,5, 0.1)
x2 = np.arange(0,5, 0.2)
```

```python
plt.figure(1)
#figure es opcional si haces una figura conjunta
plt.subplot(211)
#subplot es para especificar los ejes se usan 2 filas y una columna, 
#2= dos filas
#1= una columna
#1= primera figura
plt.plot(x1, f(x1),'ro',x2, f(x2),'k')

plt.subplot(212)
#2= dos filas
#1= una columna
#1= segunda figura figura
plt.plot(x2, f(x2),'g--')

plt.show()
```

# Textos en los graficos

```python
mu  = 100
sigma = 20

x = mu + sigma * np.random.randn(10000)
```

```python
import matplotlib
n, bins, patches = plt.hist(x, 50, facecolor="g",alpha=0.6,)
plt.xlabel("Cociente intelectual",fontsize=14, color= "g")
plt.ylabel("Probabilidad")
plt.title(r"Histograma de CI $N\ (\mu,\sigma)$")
plt.text(40,400, r'$\mu = 100,\ \sigma=20$')
plt.axis([20,180,0,700])
plt.show()
```

```python
plt.figure(figsize=(10,6),dpi = 90)
plt.subplot(1,1,1)

x = np.arange(0,10*np.pi, 0.01)
y = np.cos(x)
plt.plot(x,y, lw=2.0)

plt.annotate('Maximo local', xy=(4*np.pi,1), xytext = (15, 1.5), arrowprops=dict(facecolor = "black",shrink=0.05))

plt.show()
#figsize me da el alto y largo de la figura
#dpi calidad de grafico
#lw tamaño de linea
#annotate le integra un texto al grafico
#xytext las coordenadas del texto
#arrowprops pone una flecha
#shrink el tamaño de la flecha
```

```python
from matplotlib.ticker import NullFormatter
```

```python
mu = 0.5
sigma = 0.3

y = mu +sigma*np.random.randn(1000)
y = y[(y>0) & (y<1)]
y.sort()
x = np.arange(len(y))
```

```python
plt.figure(1)

plt.subplot(2,2,1)
plt.plot(x,y)
plt.yscale("linear")
plt.xscale("linear")
plt.title("Escala lineal")

plt.subplot(2,2,2)
plt.plot(x,y)
plt.yscale("log")
plt.title("Escala logaritmica")

plt.subplot(2,2,3)
plt.plot(x,y - y.mean())
plt.yscale("symlog", linthreshy=0.01)
plt.title("Escala log simetrico")

plt.subplot(2,2,4)
plt.plot(x,y)
plt.yscale("logit")
plt.title("Escala logistica")
plt.gca().yaxis.set_minor_formatter(NullFormatter()) #esto es para eliminar formatos del eje y

plt.subplots_adjust(top=0.92, bottom = 0.08, left =0.1, right =0.95, hspace=0.6, wspace=0.50)
# con eso acomodas las figuras, hspace alto y wespace el ancho
plt.show()
```
